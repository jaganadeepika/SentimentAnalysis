```python
import os
import re
import joblib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.utils import resample
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
RANDOM_STATE = 42
```

```python
df = pd.read_csv("Datafiniti_Amazon_Reviews.csv")  # change path if needed
print("Rows, Columns:", df.shape)
display(df.head(3))
null_counts = df.isnull().sum().sort_values(ascending=False)
print("\nTop missing-value columns:\n", null_counts[null_counts > 0].head(20))
```

```python
print("reviews.text missing:", df['reviews.text'].isnull().sum())
print("reviews.rating missing:", df['reviews.rating'].isnull().sum())
df.dropna(subset=['reviews.text', 'reviews.rating'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
import pandas as pd
df = pd.read_csv("Datafiniti_Amazon_Reviews.csv")
df_filtered = df[["reviews.text", "reviews.rating", "name", "brand", "reviews.date"]]
df_filtered.head()
```

```python
nltk_stopwords = set(stopwords.words('english'))
negation_words = {"not", "no", "nor", "n't", "don't", "dont", "didn't", "didnt",
                  "couldn't", "couldnt", "wouldn't", "wouldnt", "won't", "wont",
                  "can't", "cant", "shouldn't", "shouldnt", "never", "none"}
custom_stopwords = nltk_stopwords - negation_words
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    """Lowercase, strip URLs/mentions, keep apostrophes, remove other punctuation,
       remove stopwords (but keep negations), and lemmatize."""
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|@\w+|#\w+", " ", text)
    text = re.sub(r"[^a-zA-Z0-9'\s]", " ", text)
    tokens = text.split()
    tokens = [tok for tok in tokens if tok not in custom_stopwords]
    tokens = [lemmatizer.lemmatize(tok) for tok in tokens]
    return " ".join(tokens)
df['cleaned_review'] = df['reviews.text'].apply(clean_text)
```

```python
df['sentiment'] = df['reviews.rating'].apply(lambda x: 'negative' if x in [1, 2] else 'positive')
df['label'] = df['sentiment'].map({'negative': 0, 'positive': 1})

print(df['sentiment'].value_counts())
```

```python
plt.figure()
sns.countplot(data=df, x='sentiment', palette='Set2')
plt.title("Sentiment Count")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()

plt.figure()
sns.countplot(data=df, x='reviews.rating', palette='viridis')
plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
from sklearn.utils import resample
df_negative = df[df['label'] == 0]
df_positive = df[df['label'] == 1]

if len(df_positive) > len(df_negative):
    df_positive_down = resample(df_positive,
                                replace=False,
                                n_samples=len(df_negative),
                                random_state=RANDOM_STATE)
    df_balanced = pd.concat([df_negative, df_positive_down])
else:
    df_negative_down = resample(df_negative,
                                replace=False,
                                n_samples=len(df_positive),
                                random_state=RANDOM_STATE)
    df_balanced = pd.concat([df_positive, df_negative_down])

df_balanced = df_balanced.sample(frac=1, random_state=RANDOM_STATE).reset_index(drop=True)
print("Balanced dataset shape:", df_balanced.shape)
print(df_balanced['sentiment'].value_counts())
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Separate positive and negative reviews
positive_reviews = df[df['sentiment'] == 'positive']['cleaned_review']
negative_reviews = df[df['sentiment'] == 'negative']['cleaned_review']

# Combine all reviews in each category
positive_text = " ".join(positive_reviews)
negative_text = " ".join(negative_reviews)

# Generate word clouds
positive_wc = WordCloud(width=800, height=400, background_color='white', colormap='Greens').generate(positive_text)
negative_wc = WordCloud(width=800, height=400, background_color='white', colormap='Reds').generate(negative_text)

# Plot both
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.imshow(positive_wc, interpolation='bilinear')
plt.axis("off")
plt.title("Positive Reviews Word Cloud")

plt.subplot(1, 2, 2)
plt.imshow(negative_wc, interpolation='bilinear')
plt.axis("off")
plt.title("Negative Reviews Word Cloud")

plt.tight_layout()
plt.show()
```

```python
X = df_balanced['cleaned_review']
y = df_balanced['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE
)

tfidf = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X_train_tfidf = tfidf.fit_transform(X_train)
X_test_tfidf = tfidf.transform(X_test)

print("TF-IDF shape (train):", X_train_tfidf.shape)
```

```python
def evaluate_model(model, X_test_vec, y_test, model_name="Model"):
    y_pred = model.predict(X_test_vec)
    acc = accuracy_score(y_test, y_pred)
    print(f"\n=== {model_name} Results ===")
    print("Accuracy:", acc)
    print("Classification Report:\n", classification_report(y_test, y_pred, digits=4))
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['neg','pos'], yticklabels=['neg','pos'])
    plt.title(f"{model_name} - Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    return acc, y_pred
```

```python
lr = LogisticRegression(max_iter=1000, random_state=RANDOM_STATE)
lr.fit(X_train_tfidf, y_train)
lr_acc, lr_pred = evaluate_model(lr, X_test_tfidf, y_test, model_name="Logistic Regression")
```

```python
nb = MultinomialNB()
nb.fit(X_train_tfidf, y_train)
nb_acc, nb_pred = evaluate_model(nb, X_test_tfidf, y_test, model_name="Multinomial Naive Bayes")
```

```python
svc = LinearSVC(random_state=RANDOM_STATE, max_iter=5000)
svc.fit(X_train_tfidf, y_train)
svc_acc, svc_pred = evaluate_model(svc, X_test_tfidf, y_test, model_name="Linear SVC")
```

```python
results_df = pd.DataFrame({
    "Model": ["LogisticRegression", "MultinomialNB", "LinearSVC"],
    "Accuracy": [lr_acc, nb_acc, svc_acc]
})
display(results_df)
sns.barplot(data=results_df, x='Model', y='Accuracy', palette='Set2')
plt.ylim(0, 1)
plt.title("Model Accuracy Comparison")
plt.show()
```

```python

feature_names = tfidf.get_feature_names_out()
weights = np.asarray(X_train_tfidf.sum(axis=0)).ravel()
top_idx = np.argsort(weights)[-20:][::-1]
top_features = [(feature_names[i], weights[i]) for i in top_idx]
top_df = pd.DataFrame(top_features, columns=['feature', 'weight'])
display(top_df.head(10))

# plot top 10
plt.figure(figsize=(10,4))
plt.bar(top_df['feature'].head(10), top_df['weight'].head(10))
plt.xticks(rotation=45)
plt.title("Top 10 TF-IDF terms across training set")
plt.show()
```

```python
joblib.dump(svc, "svm_model.pkl")
joblib.dump(tfidf, "tfidf_vectorizer.pkl")
print("Saved svm_model.pkl and tfidf_vectorizer.pkl")
```
